# Bước 1: Business Understanding & Analytic Approach
## UK Weather & Health: Weekly Deaths and Temperature (2016–2024)
---

## 1.1 Mục tiêu chính

Phân tích và dự báo mối tương quan giữa các đợt thời tiết cực đoan (nắng nóng / giá rét) và số lượng ca tử vong hàng tuần (**weekly deaths**) tại Anh.

- **Đối tượng nghiên cứu:** Toàn bộ dân cư tại khu vực **England & Wales** (dữ liệu ONS).
- **Kỳ vọng:** Tìm thấy mối quan hệ **phi tuyến** giữa nhiệt độ và tử vong — nhiệt độ quá cao hoặc quá thấp đều làm tăng số ca tử vong.

## 1.2 Định nghĩa khái niệm

| Khái niệm | Ý nghĩa | Biến đại diện |
|---|---|---|
| **Physical Health (Weather)** | Đại diện bởi nhiệt độ tối đa | `tasmax_mean_weekly_degC` |
| **Mortality (Health outcome)** | Số lượng ca tử vong thực tế ghi nhận hàng tuần | `deaths_total` |

## 1.3 ScoreGroup (0/1) — Biến `IsSurge`

Trong dự án này, ta sẽ tạo biến phân loại **`IsSurge`**:

| Giá trị | Điều kiện | Ý nghĩa |
|---|---|---|
| **1** | `deaths_total` > ngưỡng **75th percentile** | Đợt bùng phát tử vong (surge) |
| **0** | `deaths_total` ≤ 75th percentile | Mức bình thường |

Biến này sẽ được sử dụng làm **target** cho bài toán **Classification** ở các bước sau.

## 1.4 Vai trò biến

```

                        INPUT (Features)                      
  • tasmax_mean_weekly_degC  — Nhiệt độ trung bình tuần      
  • tasmax_max_weekly_degC   — Nhiệt độ cao nhất tuần        

                       OUTPUT (Targets)                       
  • deaths_total             — Số ca tử vong (Regression)     
  • IsSurge (0/1)            — Đợt bùng phát (Classification) 

```

## 1.5 Kỹ thuật phân tích

| # | Kỹ thuật | Mục đích | Output |
|---|---|---|---|
| 1 | **Tương quan (Correlation)** | Xem nhiệt độ ảnh hưởng thế nào đến tử vong | Pearson r, Spearman ρ |
| 2 | **Hồi quy (Regression)** | Dự báo số ca tử vong (`deaths_total`) | R², RMSE, MAE |
| 3 | **Phân loại (Classification)** | Dự báo các tuần có "Surge" (đột biến) — `IsSurge` | Accuracy, Precision, Recall, F1 |

## 1.6 Tổng quan Dataset

**Source file:** `uk_weather_health_weekly_2016_2024_TOTAL_FINAL.csv`

| Cột | Mô tả | Đơn vị |

## 1.7 Yếu tố nhiễu (Confounders)

 **Các đợt dịch bệnh** (như COVID-19 năm 2020) có thể làm dữ liệu tử vong tăng vọt **không do thời tiết**.

Các yếu tố nhiễu cần lưu ý:
- **Đại dịch COVID-19 (2020–2021):** Gây ra các đợt tử vong đột biến không liên quan đến nhiệt độ.
- **Mùa cúm (Flu season):** Trùng với mùa đông lạnh, gây nhiễu khi phân tích ảnh hưởng của nhiệt độ thấp.
- **Ô nhiễm không khí:** Thường tăng vào mùa đông, có thể góp phần vào tử vong nhưng không được đo trong dataset.
- **Thay đổi chính sách y tế:** Các biện pháp lockdown, tiêm vaccine có thể ảnh hưởng đến xu hướng tử vong.
